```python
#install the necessary dependencies
import pandas as pd
import matplotlib.pyplot as plt
from mplsoccer import VerticalPitch
from statsbombpy import sb
```

```python
comps = sb.competitions()
comps.columns
```

```python
# Load all event data for the match
events = sb.events(match_id= 22912)
events
```

```python
events.columns
```

```python
#only want shot events
events_table = events[["location",'type','team']]
events_table
```

```python
# Df filtering, only show shot events
shots = events[events['type'] == 'Shot']
```

```python
 #shows the shot table, the location, type and tea,

shots_table = shots[["location","type","team"]]
shots_table

 

 
```

```python
# separate dataframes, one for each team, of the shots made
spurs_shots = shots_table[shots_table["team"] == "Tottenham Hotspur"]
liverpool_shots = shots_table[shots_table["team"]=="Liverpool"]
```

```python
spurs_shots
```

```python
liverpool_shots
```

```python
#convert all liverpool shots to a list
liverpool_shots = liverpool_shots["location"].tolist()
liverpool_shots
```

```python
#convert all liverpool shots locations into separate arrays for the x and y locations, ready for use on a scatter plot
liverpool_shot_location_x = [location_x[0]  for location_x in liverpool_shots]
liverpool_shot_location_y = [location_y[1]  for location_y in liverpool_shots]
```

```python
#convert all spurs shots to a list
spurs_shots = spurs_shots["location"].tolist()
spurs_shots
```

```python
#convert all spurs shots locations into separate arrays for the x and y locations, ready for use on a scatter plot
spurs_shot_location_x = [location_x[0] for location_x in spurs_shots]
spurs_shot_location_y = [location_y[1] for location_y in spurs_shots]
spurs_shot_location_x
spurs_shot_location_y
```

```python
 
 

# Create and plot the pitch
pitch = VerticalPitch(
    pitch_type="statsbomb", 
    half=True, 
    pad_bottom=1.0,
    line_color="white", 
    pitch_color="black"
)

fig, ax = pitch.draw(figsize=(20,20))

# Scatter plot of shots
ax.scatter(spurs_shot_location_y, spurs_shot_location_x, color="white", alpha=0.9, label="Tottenham Shot Attempts", s =400)
ax.scatter(liverpool_shot_location_y, liverpool_shot_location_x, color="red", alpha=0.9, label="Liverpool Shot Attempts", s = 400)

# Add legend
ax.legend(facecolor="white", fontsize=15)

plt.show()
```
