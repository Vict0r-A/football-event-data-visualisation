```python
 #library for access StatsBomb Open Data dataset.

from statsbombpy import sb
from mplsoccer import Pitch
import matplotlib.pyplot as plt
```

```python
comps = sb.competitions()
comps.head()
```

```python
 #get the matches in this league in this season
sb.matches(competition_id=43, season_id=3)
```

```python
#all the events of a match
events = sb.events(match_id= 22912)  
events
```

```python
comps = sb.competitions() #get all the competitions
print(comps.head())
```

```python
events.columns
```

```python
#filtered df of all of the games pass events

passing_events = events[["team","type","location","pass_end_location","player"]]
passing_events = passing_events[passing_events["type"]=="Pass"]
passing_events.head(10)
```

```python
#filtered df of all Liverpools pass events
liverpool_pass_events = passing_events[passing_events['team']== "Liverpool"]
liverpool_pass_events
```

```python
#filtered df of all Spurs' pass events
spurs_pass_events = passing_events[passing_events['team']== "Tottenham Hotspur"]
spurs_pass_events
```

```python
#conversion of liverpool pass event locations into a list [x,y] elements
liverpool_start_locations = liverpool_pass_events["location"].tolist()
liverpool_start_locations #.tolist() converts column/series into a plain python list
```

```python
#stores the x coordinates and the y coordinates of the pass in separate arrays
liv_start_x = [loc[0] for loc in liverpool_start_locations]
liv_start_y = [loc[1] for loc in liverpool_start_locations]

 
```

```python
liverpool_end_locations = liverpool_pass_events["pass_end_location"].tolist()
liverpool_end_locations  
```

```python
#stores the x coordinates and the y coordinates of the pass in separate arrays
liv_end_x = [loc[0] for loc in liverpool_end_locations ]
liv_end_y = [loc[1] for loc in liverpool_end_locations ]
```

```python
#conversion of spurs pass event locations into a list [x,y] elements
spurs_start_locations = spurs_pass_events["location"].tolist()
spurs_start_locations#.tolist() converts column/series into a plain python list
```

```python
#stores the x coordinates and the y coordinates of the pass in separate arrays
spurs_start_x = [loc[0] for loc in spurs_start_locations]
spurs_start_y = [loc[1] for loc in spurs_start_locations]
```

```python
spurs_end_locations = spurs_pass_events["pass_end_location"].tolist()
spurs_end_locations  
```

```python
#stores the x coordinates and the y coordinates of the pass in separate arrays
spurs_end_x = [loc[0] for loc in spurs_end_locations ]
spurs_end_y = [loc[1] for loc in spurs_end_locations ]
```

```python
pitch = Pitch(pitch_color="green") #pitch object
fig,ax =  pitch.draw() #the matplotlib figure and axes
#figure allows control of the canvas size
#ax controls where the data is plotted
```

```python
 
#Liverpool pass map

# Create a pitch object with green background and white lines, setting the size aswell
pitch = Pitch(pitch_color="green", line_color="white")
fig, ax = pitch.draw(figsize=(10, 7))

# using blue dots to start passes location
ax.scatter(liv_start_x, liv_start_y, color="blue",  label="Pass start")

# using red dots to show end passes location
ax.scatter(liv_end_x, liv_end_y, color="red", label="Pass end")

for x1, y1, x2, y2 in zip(liv_start_x, liv_start_y, liv_end_x, liv_end_y):
    ax.annotate( "", 
        xy=(x2, y2),      
        xytext=(x1, y1),  
        arrowprops=dict(
            arrowstyle="->",
            color="black",    
        )
    )
# Create legend
ax.legend()
fig.suptitle(
    "UEFA UCL Final 2019 - Liverpool Pass Map",
    fontsize=18,
    fontweight="bold",
    color="white"
)
plt.show()
```

```python

#Spurs pass map

# Create a pitch object with green background and white lines, setting the size aswell
pitch = Pitch(pitch_color="green", line_color="white")
fig, ax = pitch.draw(figsize=(10, 7))

# using blue dots to start passes location
ax.scatter(spurs_start_x, spurs_start_y, color="blue",  label="Pass start")

# using red dots to show end passes location
ax.scatter(spurs_end_x, spurs_end_y, color="red", label="Pass end")

for x1, y1, x2, y2 in zip(spurs_start_x, spurs_start_y, spurs_end_x, spurs_end_y):
    ax.annotate( "", 
        xy=(x2, y2),      
        xytext=(x1, y1),  
        arrowprops=dict(
            arrowstyle="->",
            color="black",    
        )
    )
 

# Create legend
ax.legend()

plt.show()
```
